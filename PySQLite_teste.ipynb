```python
import sqlite3
```

```python


conn = sqlite3.connect('cyptable.db')
print("Opened database successfully");

conn.execute('''
CREATE TABLE IF NOT EXISTS Criptos(cripto text, 
                      busd integer,
                      quantidade integer, 
                      valor_de_compra integer, 
                      valor_atual integer,
liquido integer,
porcentagem integer);''')

conn.commit()

print("Table created successfully");

conn.close()
```

```python
conn = sqlite3.connect('cyptable.db')


df = pd.read_csv('resultado.csv')
for index,row in df.iterrows():
  params = row.cripto,row.busd,row.quantidade,row.valor_de_compra,row.valor_atual,row.liquido,row.porcentagem
  conn.execute("INSERT INTO Criptos VALUES (? ,? , ?, ?, ?, ?, ?)",params)

conn.commit()

print("Table created successfully");

conn.close()  
```

```python

conn = sqlite3.connect('cyptable.db')


db = conn.execute('''
SELECT * FROM Criptos''')

for row in db:
  print(row)

conn.commit()

print("Table created successfully");

conn.close()
```
